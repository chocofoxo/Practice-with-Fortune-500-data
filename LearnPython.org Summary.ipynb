```python
# Generators
# Simple functions that return an iterable set of items, one at a time
def fib():
    a, b = 1, 1
    while 1:
        yield a
        a, b = b, a + b

#Print Fibonacci sequence 10 times        
counter = 0
for n in fib():
    print(n)
    counter += 1
    if counter == 10: 
           break
```

```python
# Sets
# lists with no duplicate entries
a = set(["Jake", "John", "Eric"])
b = set(["John", "Jill"])

print(a.intersection(b)) # list of common values (John), order doesn't matter
print(a.symmetric_difference(b)) # unique values in common (Jill, Eric, Jake), order doesn't matter
print(a.difference(b)) # unique values in a that aren't in b (Jake, Eric)
print(a.union(b)) # combined set of both (no repeats)
```

```python
# Closures
# function object that remembers values in enclosing scopes even if not in memory
# data is attached to code even after function is complete
# advantage: avoid use of global variables (nonlocal) and provides some data hiding

def transmit_to_space(message):
  "This is the enclosing function"
  def data_transmitter():
      "The nested function"
      print(message)
  return data_transmitter # don't include arguments

fun2 = transmit_to_space("Burn the Sun!")
fun2() # message not used but preserved from previous function call even though function is done running

# Another example:
def multiplier_of(num):
    def multiply(x):
        print(x * num)
    return multiply

multiplywith5 = multiplier_of(5)
multiplywith5(9)
```

```python
# Decorators
# to make simple modifications to callable objects like functions, methods, or classes
# takes function and returns function

# decorator that makes function repeat twice
def repeater(old_function):
    def new_function(*args, **kwds):
        old_function(*args, **kwds) 
        old_function(*args, **kwds) # run old function twice
    return new_function

@repeater
def multiply(num1, num2):
    print(num1 * num2)

print(multiply(2, 3)) # returns 6 6 because runs twice

# modify output
def double_out(old_function):
    def new_function(*args, **kwds):
        return 2 * old_function(*args, **kwds) # modify the return value
    return new_function

# modify input
def double_Ii(old_function):
    def new_function(arg): # only works if the old function has one argument
        return old_function(arg * 2) # modify the argument passed
    return new_function

# check function
def check(old_function):
    def new_function(arg):
        if arg < 0: raise (ValueError, "Negative Argument") # raises error instead of crashing program
        old_function(arg)
    return new_function

# multiple output by a variable amount
def multiply(multiplier):
    def multiply_generator(old_function):
        def new_function(*args, **kwds):
            return multiplier * old_function(*args, **kwds)
        return new_function
    return multiply_generator # it returns the new generator

@multiply(3) # multiply is not a generator, but multiply(3) is
def return_num(num):
    return num
# Now return_num is decorated and reassigned into itself
return_num(5) # should return 15
```

```python
# Map, Filter, Reduce
# map and filter are built=in, reduce needs to be imported from functools

# Map applies function to any number of iterables (must match arguments in function)
# Returns map object, convert back with list()
circle_areas = [3.56773, 5.57668, 4.00914, 56.24241, 9.01344, 32.00013]
result = list(map(round, circle_areas, range(1, 7))) # round() needs 2 inputs, circle_areas index iterates with range

# If pass second iterable that is more/less than first, map() does as much as it can and then stops when reaches end
# Another example: zip
my_strings = ['a', 'b', 'c', 'd', 'e']
my_numbers = [1, 2, 3, 4, 5]
results = list(map(lambda x, y: (x, y), my_strings, my_numbers))

# Filter uses function that returns boolean, keeps only "true" values
# Only one iterable
dromes = ("demigod", "rewire", "madam", "freer", "anutforajaroftuna", "kiosk")
palindromes = list(filter(lambda word: word == word[::-1], dromes)) # returns palindromes only

scores = [66, 90, 68, 59, 76, 60, 88, 74, 81, 65]
def is_A_student(score):
    return score > 75
over_75 = list(filter(is_A_student, scores)) # returns scores that are over 75

# Reduce applies function of 2 args to iterable, reduces into signle value
# optional value of "initial" to serve as first arg (otherwise both first and second args must be iterable)

from functools import reduce

numbers = [3, 4, 6, 9, 34, 12]

def custom_sum(first, second):
    return first + second

result = reduce(custom_sum, numbers, 10) #starts with 10+3 -> 13, then adds 4, then 6... Result: 68+10=78
```
